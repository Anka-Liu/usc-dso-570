```python
import numpy as np
df2=df.query('year==1952')
```

```python
df2
```

```python
df2.plot(x='lifeExp',y='gdpPercap',s=np.log(df2['pop'])*5,kind='scatter')
```

```python
df.groupby('year').plot(x='lifeExp',y='gdpPercap',s=np.sqrt(df2['pop'])/100,kind='scatter')
plt.show()
```
