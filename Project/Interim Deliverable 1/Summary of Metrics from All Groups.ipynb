# Summary of Metrics from Interim Deliverable 1

There are many creative metrics among the submissions to interim deliverable 1 among both sections of DSO-570. Below is a summary of the metrics. See the actual submissions by the groups for more details. (The submissions are organized in the same directory as this Jupyter notebook, first by the section, then by the group number.)

For interim deliverable 2, since you are optimizing the schedule decision, you should focus on the metrics in the first section below, which are those that are directly affected by the scheduling.

## 1. Metrics that Can Be Directly Affected by the Scheduling Decision

### Classroom Utilization Rate

The idea is to examine the ratio between the capacity of the classroom and the number of registered students. It is desirable to use big classrooms only for courses that require many seats, otherwise we have wasted seats. Inefficiencies in classroom utilization is a resounding theme among the submissions. Groups that examine this metric includes:

- Section 1: Groups 1, 2, 3, 6, 9 and 10.
- Section 2: Groups 3, 5, 8 and 9

The downside to reducing extra seat capacity in courses is that in the event that the course is more popular than we thought, then the room will not be able to accommodate extra students. Moreover, it might be difficult to change the room at that point since other rooms may be already taken (although one may be able to run another optimization at that point to reshuffle rooms). Group 8 of Section 2 also examines the set of courses in which there are no empty seats, and uses this as an indicator that room size was a bottleneck for allowing more students to take these popular courses.

### Pairs of Courses to Avoid Scheduling Conflicts

The idea is to identify pairs of courses that should be offered at different times, so as to enable students to take both courses in the same semester. 

Group 2 of Section 2 identifies pairs of courses that should not conflict by studying the student course selection data and identifying pairs of courses that many students take together.

Group 7 of Section 2 identifies pairs of courses by finding the most popular courses and postulating that these should not conflict. (Note that to do this properly, one should filter by the program, as courses that are popular among the MBAs may not be popular among the MSBAs, or among the undergrads.)

### Student and Faculty Time Preferences

The idea is to either collect additional data from faculty or students about preferred time slots, or to postulate for example that students do not like early mornings and do not like gaps in their schedule.

Group 5 of Section 1 examines by department the percentage of course times that are outside of a certain desirable "prime time." They assume that the majority of faculty and students prefer to schedule courses between 9am and 5pm on Monday to Fridays.  

Group 1 of Section 2 seeks to identify popular times for students by looking at time slots in which classes tend to be more full. They also propose to study the student course selection data and identify whether students have gaps in their schedule for a given day. The same idea of examining gaps in the schedule can be applied to faculty as well, as many faculty prefer to teach consecutive slots rather than have gaps between teaching during the day.

Group 3 of Section 2 proposes to collect additional faculty data on time slots by asking each professor to give their three most preferred time slots, and measure whether one is able to fulfil these wishes. 

Group 6 of Section 2 postulates that preferences between full-time and part-time students are different: for full-time students, they measure maximum time spent in class on any given day of week; for part-time students, they measure the number of days of the week a student must come to school.

Group 2 of Section 1 suggests using revenue to prioritize which courses to give the best time slots (say 10am-5pm on Mondays through Fridays). Their metric essentially prioritizes courses with higher enrollment, which makes sense as it's better overall to make a few students come early or stay late than to make everyone in a large introductory class to do so.

## 2. Metrics that are Not Directly Affected by the Scheduling Decision

The following proposed metrics can be used to by Shannon and Hal as "health indicators" of the overall scheduling system. However, these metrics are not directly affected by the scheduling decision, and are perhaps less relevant for the Interim Deliverable 2. Nevertheless, you can include these and other metrics in your final project report.

### Overlap between Phase 1 Allocation and Final Schedule

Group 4 of Section 2 examines whether the allocation to each department in phase 1 matches what they get in the final schedule. In the data, there are courses from a department that are assigned to slots of other departments. This is due to the slots from the other department being unused at the end of Phase 1, so Hal takes these unused slots and assign remaining courses centrally. This group measures the amount of overlap between original slot assignment and the final course assignment, and use this as a proxy of the percentage of courses that are scheduled in Phase 1. The same methodology can also be used to estimate whether a particular course was scheduled in Phase 1 or 2 (based on whether the course is scheduled within the department's given slot or not.)

While the amount of overlap is within the control of Shannon/Hal, one would not want to schedule in such as way to stay within the current allocation of slots, as those slots may not be optimally allocated. Moreover, it might not make sense to change the current allocation of slots to better match the current schedules, as the current schedules might not be optimal. It is better to try to optimize the current schedules directly.

### Cancellation Rate of Courses

The idea here is to examine the set of courses that have been proposed (the merger of the course enrollment and the cancellation files), and calculate what percentage of proposed courses were cancelled. One can also break this down by department and identify departments with the highest cancellation rates. Groups that considered cancellation rate include groups 4 and 8 of Section 1, and groups 5 and 8 of Section 2.

Moreover, it is hard to predict how much effect changing the schedule has on the cancellation rate. In the optimization of schedules, one should focus on directly improving a metric that measures efficiency of utilization.  

As a side remark, one must be careful in interpreting high cancellation rates as bad, as a department having a high cancellation rate might simply be innovative in proposing many new courses, some of which may not be approved by the curriculum office or may not attract enough demand. On the other hand, a department that never innovates will have few course cancellation. 



### Student Faculty Ratio 

As a measure of the quality of education for students, group 7 of Section 1 examines the student-faculty ratio: the average number of students per course in Marshall. They also study this metric across programs and identify departments and programs with abnormally large sizes. 

The number of students allowed in each course is not something within the control of the scheduling system, as that is determined beforehand. Furthermore, courses in entrepreneurship (which are mainly based on case discussions) may require smaller classroom sizes than courses in DSO (which allow for larger lecture style). Hence, one must be careful in interpreting these results. However, the overall student faculty ratio can be a useful indicator of educational quality, as this metric is already used widely for this purpose. 

### Course Popularity

Groups 5 and 8 of Section 2 also consider the popularity of courses. Group 5 uses the number of students who are enrolled (big courses with many sections are popular). Group 8 also uses whether the available seats are all taken (classes that have no more empty space are popular, as this indicates the existence of other students who wanted to take the course but could not).

While one can use course popularity to prioritize which courses get the best time slots, it is hard to predict how popularity changes with a different schedule, so it may not be appropriate to use this as an objective in the optimization.

### Revenue

Group 2 of Section 1 also estimates the total revenue to Marshall from the courses. For each course, they estimate the revenue to Marshall from the course by multiplying the number of enrolled students, the number of credits, and the tuition per credit. 

While it makes sense to prioritize the highest revenue courses (by this metric) in scheduling, as that amounts to prioritizing the biggest classes, one has to be careful in using this metric to prioritize resources. For example, a specialized elective course may have few students enrolled at a given time (so is low according to this metric), but we still need to devote enough resources to specialized electives to attract people to come to Marshall in the first place. (In other words, the specialized elective classes attract students to Marshall, and the big mandotary courses collect the payments.)
