## Optimization Formulation from Lab 7
The following formulation is from Lab 7. It is reproduced for your convenience.

**Decision variables:** 

- $x_i$: the new weight of stock $i$. (Continuous)
- $z_i$: whether to give stock $i$ a positive weight. (Binary)
- $a_i$: an auxiliary variable to encode $|x_i-w_i|$. (Continuous)

**Formulation (last 3 constraints linearized):**
$$\begin{aligned}
\text{Maximize:} && \sum_{i} R_i x_i & && \text{(Average Return)}\\
\text{subject to:} \\
\text{(Valid weights)} && \sum_i x_i & = 1 \\
\text{(Risk tolerance)} && \sum_{i,j} C_{ij}x_ix_j & \le \sigma_{target}^2 \\
\text{(Change in weights)} &&  \frac{1}{2} \sum_i a_i & \le \delta \\
\text{(At most $k$ stocks)} && \sum_{i}z_i & \le k \\
\text{(Absolute value-1)} && x_i-w_i & \le a_i &&\text{for each stock $i$.}\\
\text{(Absolute value-2)} && -(x_i-w_i) & \le a_i &&\text{for each stock $i$.}\\
\text{(Stocks to include)} && x_i & \le z_i && \text{for each stock $i$.}\\
\text{(Non-negligible weights)} && x_i & \ge \epsilon z_i && \text{for each stock $i$.} \\
&& x_i, a_i & \ge 0 \\
&& z_i & \in \{0,1\} \\
\end{aligned}$$


## Gurobi Code from Lab 7

````python

import gurobipy as grb
mod=grb.Model()

S=ret.index
x={}
z={}
a={}
for i in S:
    x[i]=mod.addVar(lb=0)
    z[i]=mod.addVar(vtype=grb.GRB.BINARY)
    a[i]=mod.addVar(lb=0)

totRet=grb.quicksum(ret.loc[i]*x[i] for i in S)

mod.setObjective(totRet,sense=grb.GRB.MAXIMIZE)

totCov=grb.quicksum(cov.loc[i,j]*x[i]*x[j] for i in S for j in S)

covCap=mod.addConstr(totCov<=stdMax**2)
numCap=mod.addConstr(grb.quicksum(z[i] for i in S)<=k)
change=mod.addConstr(grb.quicksum(a[i] for i in S)/2<=maxChange)
mod.addConstr(grb.quicksum(x[i] for i in S)==1)

for i in S:
    mod.addConstr(x[i]-oldPortfolio.loc[i,'Weight']<=a[i])
    mod.addConstr(-x[i]+oldPortfolio.loc[i,'Weight']<=a[i])
    mod.addConstr(x[i]<=z[i])
    mod.addConstr(x[i]>=eps*z[i])

mod.optimize()
table=[]
for i in S:
    if x[i].x>0:
        table.append([i,x[i].x])
portfolio=pd.DataFrame(table,columns=['Stock','Weight'])
portfolio=portfolio.sort_values(by='Weight',ascending=False)
portfolio.to_excel(outputFile,index=False)
````

**Note how we can set a linear expression in Gurobi to a variable, and later refer to this using the variable.**

```python
totRet=grb.quicksum(ret.loc[i]*x[i] for i in S)
mod.setObjective(totRet,sense=grb.GRB.MAXIMIZE)
```

**After optimization, we can then get the value of this expression using `getValue()` as below.**

````python
print('Optimal portfolio')
print('\t Yearly Return', totRet.getValue())
print('\t Annual volatility (standard deviation)', np.sqrt(totCov.getValue()))
portfolio
````

## Sensitivity Analysis

**After constructing the model, Gurobi allows one to alter the right hand side (RHS) of a constraint by simply setting the `RHS` attribute of the constraint. This requires first setting the constraint to a variable, which we did above with `covCap`, `numCap` and `change`. We can also change the objective by using `mod.setObjective`.** Changing the formulation is much faster than recreating it from scratch.

### Minimum Volatility Possible 

The following computes the minimum volatility portfolio, allowing as to use every stock ($k=450$) and change the entire portfolio ($\delta=1$).

````python
table=[]
numCap.RHS=450
change.RHS=1
mod.setObjective(totCov,sense=grb.GRB.MINIMIZE)
mod.setParam('OutputFlag',False)
mod.optimize()
print('Minimum volatility possible:', np.sqrt(totCov.getValue()))
````

### Tradeoff between risk and return

The following plots the relationship between risk and return when we allow every stock ($k=450$) and any sized change ($\delta=1$). We use the min. volatility of $0.151$ in the plots (otherwise the optimization would be infeasible). As can be seen, the higher the risk tolerance, the higher the possible return. However, there is decreasing marginal returns as the slow of the graph tapers off.

````python
mod.setObjective(totRet,sense=grb.GRB.MAXIMIZE)
mod.setParam('OutputFlag',False)
numCap.RHS=450
change.RHS=1

import matplotlib.pyplot as plt
xValues=np.linspace(0.151,1,20)
yValues=[]

for std in xValues:
    covCap.QCRHS=std**2
    mod.optimize()
    yValues.append(totRet.getValue())
print(yValues)
plt.plot(xValues,yValues,'o')
plt.title('Trade-off between return and risk')
plt.xlabel('Risk tolerance (yearly standard deviation)')
plt.ylabel('Yearly return')
plt.show()
````

### How much diversification is enough?

In order to decrease risk, one has to diversify the investment. However, the following shows that one only needs to diversify to about 20 stocks, after which there is little benefit from diversification. This is why the original parameters were chosen so that $k=20$. A low number of stocks to hold also has benefit in decreasing the complexity of maintaining the portfolio. 

````python
mod.setParam('OutputFlag',False)
mod.setObjective(totCov,sense=grb.GRB.MINIMIZE)
change.RHS=1
retSeries=pd.Series()
covCap.QCRHS=1
K=range(5,101,5)
yValues=[]
for newk in K:
    numCap.RHS=newk
    mod.optimize()
    yValues.append(np.sqrt(totCov.getValue()))
print(yValues)
plt.plot(K,yValues,'o')
plt.title('Trade-off between risk and complexity of portfolio')
plt.xlabel('# of stocks allowed')
plt.ylabel('Yearly volatility')
plt.show()
````

### How much change is needed to current weights?

The following graph illustrates the relationship between the parameter $\delta$ and the optimal return. As can be seen, the optimal return is almost linear in $\delta$ up to $\delta=0.6$, after which it tapers off. In practice, it is desirable to set a small $\delta$ because large changes incurs high transaction cost. Furthermore, as new data is available, one would want to update the portfolio incrementally based on previous position, instead of having to get a completely new set of stocks.

````python
mod.setObjective(totRet,sense=grb.GRB.MAXIMIZE)
covCap.QCRHS=stdMax**2
numCap.RHS=k
xValues=np.linspace(0.2,1,11)
yValues=[]
for C in xValues:
    change.RHS=C
    mod.optimize()
    yValues.append(totRet.getValue())
print(yValues)
plt.plot(xValues,yValues,'o')
plt.title('Trade-off between return and change allowed')
plt.xlabel('Change allowed to the weights')
plt.ylabel('Yearly return')
plt.show()
````
